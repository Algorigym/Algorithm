-   ìœ„ ë¸”ë¡œê·¸ì—ì„œ heapqë¥¼ ì œëŒ€ë¡œ ì´í•´í•˜ì§€ ëª»í•´ì„œ ê°€ë…ì„± í¸ì˜ë¥¼ ìœ„í•´ `append([v, w])` ë¡œ ë°”ê¿¨ë”ë‹ˆ ì˜¤ë¥˜ê°€ ë‚¨. `# ê±°ë¦¬ë¥¼ ì•ì—ë‘ì–´ ìš°ì„ ìˆœìœ„ íì— ë„£ì„ë•Œ ê±°ë¦¬ê°€ ë¹„êµë˜ë„ë¡í•œë‹¤.` ì£¼ëª©! => heapq ë‹¤ì‹œ í•œ ë²ˆ ê°œë… ë³µìŠµí•˜ê¸°



```python

import sys
import heapq

# ip = input # ì£¼í”¼í„°ì—ì„œëŠ” readline ì•ˆë¨¹í˜€ì„œ input ì¨ì•¼ í•¨.
ip = sys.stdin.readline # ì œì¶œí•  ë•Œë§Œ ì£¼ì„ í•´ì œ. ì‹œê°„ì´ ë‹¨ì¶•ë˜ë¯€ë¡œ. (1200ms to 868ms)

INF = sys.maxsize
V, E = map(int, ip().split())

G = [[] for _ in range(V + 1)]
K = int(ip())

for _ in range(E):
    u, v, w = map(int, ip().split())
    G[u].append([w, v])

result = [INF for _ in range(V + 1)]
result[K] = 0

q = []
heapq.heappush(q, [0, K]) # ê±°ë¦¬ë¥¼ ì•ì—ë‘ì–´ ìš°ì„ ìˆœìœ„ íì— ë„£ì„ë•Œ ê±°ë¦¬ê°€ ë¹„êµë˜ë„ë¡í•œë‹¤.

while q:
    dist, to = heapq.heappop(q)

    if result[to] < dist: # í•´ë‹¹ ê²½ë¡œê°€ ì´ì „ê²½ë¡œë³´ë‹¤ ê¸¸ë‹¤ë©´ ì—°ê²°ëœ ë…¸ë“œì— ëŒ€í•´ íƒìƒ‰í• í•„ìš”ê°€ì—†ë‹¤.
        continue
    for d, t in G[to]: # ì—°ê²°ëœ ë…¸ë“œ íƒìƒ‰
        d += dist # ì´ì „ê±°ë¦¬ì™€ í˜„ì¬ ì—°ê²°ëœ ë…¸ë“œì˜ ê±°ë¦¬ë¥¼ ë”í•´ì„œ
        if d < result[t]: # ê±°ë¦¬ë¹„êµ # ê±°ë¦¬ê°€ ì´ì „ë³´ë‹¤ ì§§ìœ¼ë©´
            result[t] = d # ê±°ë¦¬ë¥¼ ê°±ì‹ ì‹œí‚¤ê³ 
            heapq.heappush(q, [d, t]) # ìš°ì„ ìˆœìœ„ íì— ë„£ì–´ì¤€ë‹¤.

for i in range(1, V + 1):
    print(result[i] if result[i] != INF else "INF")

```
