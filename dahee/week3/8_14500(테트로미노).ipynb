````python
tetromino = [
  {
      "rc": (1, 4),
      "directions": [(0, 0), (0, 1), (0, 2), (0, 3)] # ----
  },
  {
      "rc": (4, 1),
      "directions": [(0, 0), (1, 0), (2, 0), (3, 0)] # ----
  },
  {
      "rc": (2, 2),
      "directions": [(0, 0), (0, 1), (1, 0), (1, 1)], # 정사각형
  },
  {
      "rc": (3, 2),
      "directions": [(0, 0), (1, 0), (2, 0), (2, 1)], # ㄴ
  },
  {
      "rc": (2, 3),
      "directions": [(0, 0), (1, 0), (0, 1), (0, 2)], # 
  },
  {
      "rc": (3, 2),
      "directions": [(0, 0), (0, 1), (1, 1), (2, 1)], # 
  },
  {
      "rc": (2, 3),
      "directions": [(1, 0), (1, 1), (1, 2), (0, 2)], # 
  },
  {
      "rc": (3, 2),
      "directions": [(0, 1), (1, 1), (2, 1), (2, 0)], # ㄴ 반대
  },
  {
      "rc": (2, 3),
      "directions": [(0, 0), (1, 0), (1, 1), (1, 2)], # 
  },
  {
      "rc": (3, 2),
      "directions": [(0, 0), (1, 0), (2, 0), (0, 1)], # 
  },
  {
      "rc": (2, 3),
      "directions": [(0, 0), (0, 1), (0, 2), (1, 2)], # 
  },
  {
      "rc": (3, 2),
      "directions": [(0, 0), (1, 0), (1, 1), (2, 1)], # ㄴㄱ
  },
  {
      "rc": (2, 3),
      "directions": [(1, 0), (1, 1), (0, 1), (0, 2)], #
  },
  {
      "rc": (3, 2),
      "directions": [(1, 0), (2, 0), (0, 1), (1, 1)], # ㄴㄱ 반대
  },
  {
      "rc": (2, 3),
      "directions": [(0, 0), (0, 1), (1, 1), (1, 2)], #
  },
  {
      "rc": (2, 3),
      "directions": [(0, 0), (0, 1), (0, 2), (1, 1)], # ㅜ
  },
  {
      "rc": (3, 2),
      "directions": [(0, 1), (1, 0), (1, 1), (2, 1)], # ㅓ
  },
  {
      "rc": (2, 3),
      "directions": [(1, 0), (1, 1), (1, 2), (0, 1)], # ㅗ
  },
  {
      "rc": (3, 2),
      "directions": [(0, 0), (1, 0), (2, 0), (1, 1)], # ㅏ
  }
]

N, M = map(int, input().split())
box = []
for _ in range(N):
    box.append( list( map(int, input().split()) ) )

# s = set([])
_max = -1
for i in range(N):
    for j in range(M):
        for t in tetromino:
            # 범위 안에 있으면
            r, c = t['rc']
            if 0 <= i + r - 1 < N and 0 <= j + c - 1 < M:
                sum = 0
                for x, y in t['directions']:
                    dx, dy = i + x, j + y
                    sum += box[dx][dy]
                if _max < sum:
                    _max = sum
#                 s.add(sum)
                                   
print(_max)
````
