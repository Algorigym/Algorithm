### 알게된 것 
- 반례 대입해보기
- copy.deepcopy(list)
- dfs(재귀)는 bfs(반복문)보다 느리다.
- 경우의 수를 모두 생각하는 연습 기르기.

### 문제 해결
- 먼저 1년치를 녹인다. 덩어리를 계산한다.
- 2개로 안나눠져있으면 또 1년치를 녹인다. 덩어리를 계산한다.
- 2개로 나뉘면 끝낸다. 끝까지 2개로 안나뉘면 0을 출력한다.
    - 해결 방식은 맞았는데, 런타임 에러가 날 괴롭혔다. ```visited = [[0] * M for _ in range(N)]``` M, N 위치 헷갈려서...

- 런타임 에러 해결 후는 '틀렸습니다'가 날 괴롭혔다.
    - 첫 번째부터 2덩이로 나누어져 있는 경우를 고려 안했다. 
    - 다 녹았는데 2덩이 이상으로 안나눠져 있을 때를 검사하지 않았다.


````python
## 제출
import copy
from collections import deque

N, M = map(int, input().split())
ice = [[0] * 300 for _ in range(300)]
result = [[0] * 300 for _ in range(300)]
for i in range(N):
    aa = list( map(int, input().split()) )
    for j in range(len(aa)):
        ice[i][j] = aa[j]
        result[i][j] = aa[j]
    
year = 0

directions = [(0, -1), (0, 1), (-1, 0), (1, 0)] # 상하좌우

def getIsAllMelted(arr):
    for i in range(N):
        for j in range(M):
            if arr[i][j] > 0:
                return False
    return True

def bfs(a, b, visited):
    q = deque()
    q.append((a, b))
    visited[a][b] = 1
    
    while q:
        i, j = q.popleft()
        for x, y in directions:
            dx, dy = i + x, j + y
            if not 0 <= dx < N or not 0 <= dy < M:
                continue
            if result[dx][dy] > 0 and visited[dx][dy] == 0:
                visited[dx][dy] = 1
                q.append((dx, dy))

                
isFirst = True
while True:
    count = 0
    visited = [[0] * M for _ in range(N)]
    
    if not isFirst:
        for i in range(1, N - 1):
            for j in range(1, M - 1):
                if (ice[i][j] > 0):
                    # 0보다 크면 상하좌우 비교한다.
                    for x, y in directions:
                        dx, dy = i + x, j + y
                        if ice[dx][dy] == 0 and result[i][j] > 0:
                            result[i][j] -= 1
                            
    for i in range(1, N - 1):
        for j in range(1, M - 1):
            if result[i][j] > 0 and visited[i][j] == 0:
                count += 1
                bfs(i, j, visited)
    if not isFirst:
        year += 1
        ice = copy.deepcopy(result)
    isFirst = False
    
    # 다 녹았는지 확인, 다 녹았는데 끝까지 2덩이로 안녹으면
    isAllMelted = getIsAllMelted(result)
    if isAllMelted and count < 2:
        year = 0
    
    if count == 0:
        break
    if count >= 2:
        break
    
print(year)

````
