- 런타임 에러 (블로그랑도 로직이 같은데...!!!! 왜...!!!)

### 알게된 것 
- copy.deepcopy(list)
- dfs(재귀)는 bfs(반복문)보다 느리다.

### 문제 해결
- 먼저 1년치를 녹인다. 덩어리를 계산한다.
- 2개로 안나눠져있으면 또 1년치를 녹인다. 덩어리를 계산한다.
- 2개로 나뉘면 끝낸다. 끝까지 2개로 안나뉘면 0을 출력한다.
=> 해결 방식은 맞았는데, 런타임 에러가 날 괴롭혔다.

```python
## 제출
import copy
from collections import deque

N, M = map(int, input().split())
ice = []
for i in range(N):
    ice.append( list( map(int, input().split()) ) )
year = 0

directions = [(0, -1), (0, 1), (-1, 0), (1, 0)] # 상하좌우

def bfs(a, b, visited):
    q = deque()
    q.append([a, b])
    visited[a][b] == 1
    
    while q:
        [i, j] = q.popleft()
        
        for x, y in directions:
            dx, dy = i + x, j + y
            if not 0 <= dx < N or not 0 <= dy < M:
                continue
            if result[dx][dy] > 0 and visited[dx][dy] == 0:
                visited[dx][dy] = 1
                q.append((dx, dy))

while True:
    count = 0
    visited = [[0] * N for _ in range(M)]
    
    for i in range(1, N - 1):
        for j in range(1, M - 1):
            if (ice[i][j] > 0):
                # 0보다 크면 상하좌우 비교한다.
                for x, y in directions:
                    dx, dy = i + x, j + y
                    if ice[dx][dy] == 0 and result[i][j] > 0:
                        result[i][j] -= 1
    for i in range(1, N - 1):
        for j in range(1, M - 1):
            if result[i][j] > 0 and visited[i][j] == 0:
                count += 1
                bfs(i, j, visited)

    year += 1
    ice = copy.deepcopy(result)
    
    if count == 0:
        break
    if count >= 2:
        break
    
print(year)
```
